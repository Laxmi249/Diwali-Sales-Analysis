# Seaborn Library

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
var1=[1,2,3,4,5,6,7,8]
var2=[8,7,6,5,4,3,2,1]  #make datasets
```

```python
plt.plot(var1,var2)  #Using matplotlib 
plt.show()
```

```python
sns.lineplot(x=var1,y=var2)  #Using seaborn
plt.show()
```

```python
var1=[1,2,3,4,5,6,7,8]
var2=[8,7,6,5,4,3,2,1]
import pandas as pd
df=pd.DataFrame({'var1':var1,'var2':var2})  # make dataframe with dectionary key:value pair
sns.lineplot(x='var1',y='var2',data=df)   #here 'data=df' used to tell the lineplot function from where data had came,so data came from df.
plt.show()
```

In above graph we can see that the additional details like xlabel and ylabel adds by default using seaborn library.

```python
df1=sns.load_dataset("penguins")  #Load a seaborn dataset
```

In above penguins dataset loaded from 'https://github.com/mwaskom/seaborn-data/blob/master/penguins.csv',we directly give the dataset name in attribute. 

```python
df1
```

```python
# plot a lineplot with bill_length in x-axis and flipper_length in y-axis
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex',palette='rocket_r')  # pallet used to give the different color 
plt.show()
```

```python
#to give the marker dot to the male and traingle to the female.
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex',palette='rocket_r',style='sex',markers=["o",">"]) 
plt.show()
```

# Bar Plot In Seaborn

```python
df1=sns.load_dataset("penguins")
```

```python
df1
```

```python
df1['island']
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm)   # Bar graph with island in x-axis and bill_length in y-axis
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1)   #Alternate way
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex') #divide in terms of gender  
plt.show()
```

```python
order1=[ 'Dream','Torgersen','Biscoe island',]
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',order=order1)  #change the order of island 
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',hue_order=['Female','Male']) #change the order of gender 
plt.show()
```

```python

sns.barplot(x='island',y='bill_length_mm',data=df1,color='r') #change the color
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',hue_order=['Female','Male'],palette='Accent') #change the color of gender 
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',hue_order=['Female','Male'],palette='prism') #change the color of gender 
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',hue_order=['Female','Male'],palette='gist_heat_r') #change the color of gender 
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',hue_order=['Female','Male'],palette='gist_heat_r',saturation=0.5) #change the  fade of color of gender use saturation parameter 
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df1,hue='sex',hue_order=['Female','Male'],palette='gist_heat_r',saturation=0) #change the  fade of color of gender 
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'])  #make histogram graph with count of flitter length
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,180,200,210,220,230,240]) # create graph with interval
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,180,200,210,220,230,240],color='g') # change color
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],bins=[170,180,180,200,210,220,230,240],kde=True )  # kde use for kernel density
plt.show()
```
